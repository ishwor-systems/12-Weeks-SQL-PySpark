1: Latest Transaction per User
You have table transactions with column name – user_id(int), txn_date (date) and amount(int)
Table structure:
(1, '2024-01-01', 100),
(1, '2024-02-01', 200),
(2, '2024-01-05', 300)


```python
%sql
CREATE TABLE transactions (
    user_id   INT,
    txn_date  DATE,
    amount    INT
);

INSERT INTO transactions (user_id, txn_date, amount) VALUES
(1, '2024-01-01', 100),
(1, '2024-02-01', 200),
(2, '2024-01-05', 300);
```

```python
%sql
---Problem-Find Latest Transcation for Each User
WITH ranked_txns AS (
    SELECT
        user_id,
        txn_date,
        amount,
        ROW_NUMBER() OVER (
            PARTITION BY user_id
            ORDER BY txn_date DESC
        ) AS rn
    FROM transactions
)
SELECT user_id, txn_date, amount
FROM ranked_txns
WHERE rn = 1;
```
