CASE STUDY 4: 
Funnel Analysis
You have table funnel with column names – user_id and step (e.g., view, cart and purchase)
Table structure:
(1, 'view'),
(1, 'cart'),
(1, 'purchase'),
(2, 'view'),
(2, 'cart')
Problem – Count users who completed all steps (view -> cart -> purchase)


```python
%sql
CREATE TABLE funnel (
    user_id  INT,
    step     VARCHAR(20)
);

INSERT INTO funnel (user_id, step) VALUES
(1, 'view'),
(1, 'cart'),
(1, 'purchase'),
(2, 'view'),
(2, 'cart');
```

```python
%sql
--Problem – Count users who completed all steps (view -> cart -> purchase)
SELECT COUNT(*) AS users_completed_all
FROM (
    SELECT user_id
    FROM funnel
    WHERE step IN ('view', 'cart', 'purchase')
    GROUP BY user_id
    HAVING COUNT(DISTINCT step) = 3
) AS completed_users;
```
