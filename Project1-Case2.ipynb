CASE STUDY 2: 
1: Top Product per Category
You have a table name is sales with column name category, product and revenue
Table structure:
('Electronics', 'Phone', 1000),
('Electronics', 'Laptop', 2000),
('Electronics', 'Tablet', 1500),
('Clothing', 'Shirt', 500),
('Clothing', 'Jacket', 800)
Problem – Find highest selling product in each category.


```python
%sql
CREATE TABLE sales (
    category  VARCHAR(50),
    product   VARCHAR(50),
    revenue   INT
);

INSERT INTO sales (category, product, revenue) VALUES
('Electronics', 'Phone',   1000),
('Electronics', 'Laptop',  2000),
('Electronics', 'Tablet',  1500),
('Clothing',    'Shirt',    500),
('Clothing',    'Jacket',   800);
```

```python
%sql
--FInding the top selling product per category
SELECT 
    category,
    product,
    revenue
FROM (
    SELECT 
        category,
        product,
        revenue,
        RANK() OVER (
            PARTITION BY category 
            ORDER BY revenue DESC
        ) AS rnk
    FROM sales
) ranked_sales
WHERE rnk = 1;
```
